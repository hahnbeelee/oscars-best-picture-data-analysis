# Movie Gross Data Scraping

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
highest_grossing=pd.read_csv('raw data\Highest-grossing films in the United States by year.csv',delimiter=',',encoding='utf-8')
```

```python
highest_grossing.head()
```

```python
int(highest_grossing.iloc[0]['Gross'][1:].replace(',', ''))
```

```python
# changing the gross value to an int
int_grosses = []
for i, movie in highest_grossing.iterrows():
    int_gross = int(movie['Gross'][1:].replace(',', ''))
    int_grosses.append(int_gross)
```

```python
highest_grossing['Gross'] = int_grosses
```

```python
#overwrite the csv with the grosses as ints
highest_grossing.to_csv('raw data\Highest-grossing films in the United States by year.csv', index=False)
```

**Finding the gross of best picture winners**

```python
bp_wins=pd.read_csv('best_picture_winners.csv',delimiter=',',encoding='utf-8')
```

```python
og_url = 'https://www.the-numbers.com/custom-search?searchterm=&searchtype=simple'
index = og_url.find('&')
```

```python
grosses = []
for i, row in bp_wins.iterrows():
    url = og_url[:index] + row.primaryTitle + og_url[index:]
    r = requests.get(url)
    if r.status_code == 200:
        soup: BeautifulSoup = BeautifulSoup(r.content, "html.parser")
        tables = soup.findAll('table')
        if len(tables) >= 2:
            table = tables[1]
            year = table.findAll("td")[1].get_text()[-4:]
            if year == row.startYear:
                gross = int(table.findAll("td")[3].get_text()[1:].replace(',', ''))
                grosses.append(gross)
```

```python
grosses = []
for i, row in bp_wins.iterrows():
    url = og_url[:index] + row.primaryTitle + og_url[index:]
    r = requests.get(url)
    if r.status_code == 200:
        soup: BeautifulSoup = BeautifulSoup(r.content, "html.parser")
        tables = soup.findAll('table')
        if len(tables) >= 2:
            table = tables[1]
            year = table.findAll("td")[1].get_text()[-4:]
            if year == str(row.startYear):
                gross = int(table.findAll("td")[3].get_text()[1:].replace(',', ''))
                grosses.append(gross)
            else:
                grosses.append(None)
        else:
            grosses.append(None)
    else:
        grosses.append(None)
```

```python
bp_wins['gross'] = grosses
```
